#### RQ 1: How have population demographics by age group changed over the years in the dataset? 

```python
directory_name = "C:/Users/Jishan/Downloads/German-population-demographics-analysis-a-study-of-age-distribution/results/plots"
os.makedirs(directory_name, exist_ok=True)
```

```python
plt.figure(figsize=(16, 8))

# Create an axis for the individual age groups
ax1 = plt.gca()  
ax2 = ax1.twinx()  

age_groups = df.columns[:-1]  
for column in age_groups:
    ax1.plot(df['Year'], df[column], label=column)

ax2.plot(df['Year'], df['Total'], 'k-', label='Total', linewidth=2)

# Setting labels
ax1.set_xlabel('Year')
ax1.set_ylabel('Age Group Population')
ax2.set_ylabel('Total Population')

# Setting title and grid
plt.title('Population Demographics by Age Group Over the Years')
ax1.grid(True)

# Handling legends - set a smaller font size
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='lower right', fontsize='8') 

plt.tight_layout()
plt.savefig('C:/Users/Jishan/Downloads/German-population-demographics-analysis-a-study-of-age-distribution/results/plots/Population_Demographics_by_Age_Group.png')
plt.show()
```

#### Interpretation

The graph "Population Demographics by Age Group Over the Years" from around 1970 to 2020 illustrates various demographic shifts. Notably, there's an overall increase in the total population, as shown by the black line, with a significant spike starting just before 2000. The age group lines display cyclical trends suggesting fluctuations in birth rates. In recent years, there is a visible aging trend, with lines representing older age groups (closer to 19 years) maintaining higher positions, indicating fewer young children and potentially lower fertility rates. This spread in population across different age groups in the later years also hints at demographic changes such as extended life expectancy.

#### RQ 2: In which years was there a significant change in the age distribution?

```python
# Selecting a few age groups for simplicity
age_groups = ['under 1 year', '5 years', '10 years', '15 years', '19 years']

# Plotting population trends by age group
plt.figure(figsize=(14, 8))
for age_group in age_groups:
    plt.plot(df['Year'], df[age_group], label=age_group)
plt.title('Population Trends by Age Group')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend()
plt.grid(True)
plt.savefig('C:/Users/Jishan/Downloads/German-population-demographics-analysis-a-study-of-age-distribution/results/plots/Age_Distribution_Analysis.png')
plt.show()
```

#### Interpretation

The graph titled "Population Trends by Age Group" illustrates the fluctuations in population for different age groups from around 1970 to 2020. Each line represents an age group: under 1 year, 5 years, 10 years, 15 years, and 19 years. Key observations include notable volatility and a general declining trend in population for younger age groups beginning around 2000. There's also a marked recovery across all age groups during the 1990s, possibly due to increased birth rates or changes in demographic policies. Towards 2020, the population numbers for these groups converge, suggesting a stabilization in the distribution across these cohorts, reflecting broader demographic shifts and possible societal changes.

```python
changes_df = df.set_index('Year').pct_change().reset_index()

# Plotting the percentage changes for the same selected age groups
plt.figure(figsize=(14, 8))
for age_group in age_groups:
    plt.plot(changes_df['Year'], changes_df[age_group] * 100, label=f'% Change in {age_group}')  
plt.title('Year-over-Year Percentage Change in Population by Age Group')
plt.xlabel('Year')
plt.ylabel('Percentage Change')
plt.legend()
plt.grid(True)
plt.savefig('C:/Users/Jishan/Downloads/German-population-demographics-analysis-a-study-of-age-distribution/results/plots/Age_Distribution_Analysis_1.png')
plt.show()
```

#### Interpretation

The graph "Year-over-Year Percentage Change in Population by Age Group" shows the annual percentage changes in population for various age groups from 1970 to around 2020. There is significant fluctuation in the population changes across all age groups, with notable spikes around the year 2000, indicating dramatic increases in population percentages, possibly due to specific events or policy changes that year. After these spikes, the percentage changes normalize and remain mostly within a narrower range of fluctuation. The general trend in more recent years shows moderate variability but no extreme changes, suggesting a stabilization of population growth rates in these age groups.

```python
percentage_changes = {}
for column in df.columns[1:-1]: 
    initial_value = df[column].iloc[0]
    percentage_changes[column] = (df[column].iloc[-1] - initial_value) / initial_value * 100

# Set a threshold for significant change
threshold = 5  # I choose here 5

significant_changes = [year for year, change in percentage_changes.items() if abs(change) > threshold]
print("Years with significant changes in the age distribution (threshold: 5):", significant_changes)
```

#### Interpretation

The output highlights years where significant changes in age distribution occurred, with changes exceeding a 5% threshold value across various age groups from "under 1 year" up to "16 years". This result suggests that during these years, there were notable demographic shifts within these age groups, possibly due to events or policies that had a substantial impact on birth rates or population movements.

#### RQ 3: How has the number under 1 year population changed (Increased/Decreased) from 1970 to 2022 in Germany?

```python
# Filter the rows for the years 1970 and 2022
population_1970 = df[df['Year'] == '31/12/1970']['under 1 year'].values[0]
population_2022 = df[df['Year'] == '31/12/2022']['under 1 year'].values[0]

# Calculate the change in population from 1970 to 2022
population_change = population_2022 - population_1970

# Determine if the population has Increased or Decreased
change_status = "Increased" if population_change > 0 else "Decreased"

# Print the change in population status
print(f"The number of individuals under 1 year population has {change_status} from 1970 to 2022 in Germany.")
```

#### Interpretation

I can observe the above output number of individuals under 1 year population has Decreased from 1970 to 2022 in Germany.

```python
absolute_change = abs(population_change)
print(f"The population has {change_status} by {absolute_change} individuals.")
```

#### Interpretation

And here the population has Decreased by 46833 individuals.

```python
# Filter the rows for the years 1970 and 2022
population_1970 = df[df['Year'] == '31/12/1970']['19 years'].values[0]
population_2022 = df[df['Year'] == '31/12/2022']['19 years'].values[0]

# Calculate the change in population from 1970 to 2022
population_change = population_2022 - population_1970

# Determine if the population has Increased or Decreased
change_status = "Increased" if population_change > 0 else "Decreased"

# Print the change in population status
print(f"The number of individuals under 19 year population has {change_status} from 1970 to 2022 in Germany.")
```

#### Interpretation

The number of individuals under 19 year population has Increased from 1970 to 2022 in Germany.

```python
absolute_change = abs(population_change)
print(f"The population has {change_status} by {absolute_change} individuals.")
```

#### Interpretation

The population has Increased by 6094 individuals.

#### RQ 4: In which year the minimum population has been found under any age group from 1970 to 2022 in Germany?

```python
# Convert the 'Year' column from string to datetime format
df['Year'] = pd.to_datetime(df['Year'], format='%d/%m/%Y')

# Calculate the minimum population for each year across all age groups
min_population_year = df.set_index('Year').drop(columns=['Total']).sum(axis=1).idxmin().year

# Print the year with the minimum population
print("The year with the minimum population under any age group from 1970 to 2022 is:", min_population_year)
```

#### Interpretation

Between the year from 1970 to 2022, the minimum population under any age group in the year of 1988.

#### RQ 5: What is the increase/decrease rate 18 years age group population from 1970 to 2022 in Germany?

```python
# Extract population values for the years 1970 and 2022 for the 18 years age group
population_1970 = df.loc[df['Year'] == '31/12/1970', '18 years'].values[0]
population_2022 = df.loc[df['Year'] == '31/12/2022', '18 years'].values[0]

# Calculate the percentage change
percentage_change = ((population_2022 - population_1970) / population_1970) * 100

# Print the percentage change
print("The percentage change in the population of the 18 years age group from 1970 to 2022 in Germany is:", percentage_change)
```

The percentage change in the population of the 18 years age group from 1970 to 2022 in Germany is: -0.8060728117738187
